```python
import pickle
import numpy as np
from tabulate import tabulate

# Specify the path to your pickle file
file_path = '/content/drive/MyDrive/Colab Notebooks/DeepDEP/premodel_tcga_exp_500_200_50.pickle'

# Open the file in binary mode
with open(file_path, 'rb') as file:
    # Load the pickle object
    data = pickle.load(file)

# Convert the data to a table format
table = []
for item in data:
    if isinstance(item, np.ndarray):
        table.append([f"Array ({item.shape})", item])
    else:
        table.append([type(item), item])

# Display the table
print(tabulate(table, headers=["Type", "Value"], tablefmt="pretty"))
```
