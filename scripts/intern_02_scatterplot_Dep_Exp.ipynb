```python
# check for any negative values for dependency

import pandas as pd

# Read in the 22Q2 CRISPR_gene_dependency.csv
dependency = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q2 CRISPR_gene_dependency.csv", index_col=0)

# Find columns with negative values
negative_columns = dependency.columns[dependency.lt(0).any()]

# Display columns with negative values
print("Columns containing negative values:")
print(negative_columns)
```

```python
# check for any negative values for expression

import pandas as pd

# Read in the 22Q2 CCLE_expression.csv
expression = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q2 CCLE_expression.csv", index_col=0)

# Find columns with negative values
negative_columns = expression.columns[expression.lt(0).any()]

# Display columns with negative values
print("Columns containing negative values:")
print(negative_columns)
```

```python
# Find the common column names
common_cols = set(dependency.columns).intersection(set(gene_expression.columns))

# Print the common column names
print(common_cols)
```

```python
# Find the number of common column names
num_common_cols = len(common_cols)
print(f"There are {num_common_cols} common column names between the two datasets.")
```

```python
# Merge the two files based on the cell line ID (after merged only keep the common row names)
merged_data = pd.merge(dependency, gene_expression, left_on="DepMap_ID", right_index=True)
merged_data
```

```python
# Get the first 10 columns of each dataset
dependency_cols = list(dependency.columns)[1:11]
gene_expr_cols = list(gene_expression.columns)[1:11]

# Print out the names of the columns
print(dependency_cols)
print(gene_expr_cols)
```

```python
print(merged_data.columns)
```

```python
# Get the column names of the merged file that contain "_x"
x_cols = [col for col in merged_data.columns if col.endswith('_x')]

# Loop through the _x columns and print their corresponding _y columns
for col in x_cols:
    matching_col = col[:-2] + '_y'
    if matching_col in merged_data.columns:
        print(f"{col} matches with {matching_col}")
```

```python
# Get the first 10 columns of merged_data
x_cols = [col for col in merged_data.columns if col.endswith('_x')]
print(x_cols [0:10])
```

```python
# Generate scatter plots for the first 10 matching column pairs

import matplotlib.pyplot as plt

# Get the column names of the merged file that contain "_x"
x_cols = [col for col in merged_data.columns if col.endswith('_x')]

# Get the corresponding matching column names that end with "_y"
matching_cols = [col[:-2] + "_y" for col in x_cols if col[:-2] + "_y" in merged_data.columns]

# Generate scatter plots for the first 10 matching column pairs
for i in range(min(len(matching_cols), 10)):
    col_x = x_cols[i]
    col_y = matching_cols[i]
    plt.scatter(merged_data[col_x], merged_data[col_y])
    plt.xlabel(col_x+  " dependency")
    plt.ylabel(col_y+  " gene_expression")
    plt.show()
```

```python
# Generate scatter plots for A1BG and A1CF
genes_to_plot = ["A1BG (1)", "A1CF (29974)"]

for gene in genes_to_plot:
    col_x = gene + "_x"
    col_y = gene + "_y"
    plt.scatter(merged_data[col_x], merged_data[col_y])
    plt.xlabel(gene +  " gene_dependency")
    plt.ylabel(gene +  " gene_expression")
    plt.show()
```
