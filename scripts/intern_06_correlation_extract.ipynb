```python
# To extract ALL values bigger than 0.1 and create a new dataset preserving the rows and columns with values bigger than 0.1 for each interest columns EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep

import pandas as pd
from IPython.display import display

# Read the CSV file into a DataFrame
correlation_table = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/correlation_table_02.csv', index_col=0)

# Define the columns you want to extract
columns_to_extract = ['EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep']

# Create a dictionary to store the extracted datasets
extracted_datasets = {}

# Iterate over each column and save them separately
for column_name in columns_to_extract:
    # Extract the column values bigger than 0.1
    column_values_gt_0_1 = correlation_table.loc[correlation_table[column_name] > 0.1]

    # Save the column values to a CSV file
    file_name = f'/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/{column_name.replace(" ", "_")}_gt_01.csv'
    column_values_gt_0_1.to_csv(file_name, index=False)

    # Store the extracted dataset in the dictionary
    extracted_datasets[column_name] = column_values_gt_0_1

    # Print the saved file name
    print(f"Saved column '{column_name}' to {file_name}")

# Display all the extracted datasets
for column_name, dataset in extracted_datasets.items():
    print(f"\nDataset for column '{column_name}':")
    display(dataset)
```

```python
## To extract ALL values bigger than 0.1 and create a new dataset ONLY preserving DepMap_ID columns with values bigger than 0.1 for each interest columns EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep
# Running correlation_table_all_gt_01.sh extract from file correlation_table.csv 1.76 GB to produce 'EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep_gt_01.csv

import pandas as pd
from IPython.display import display

# Read the CSV file into a DataFrame
correlation_table = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/correlation_table_02.csv', index_col=0)

# Define the columns you want to extract
columns_to_extract = ['EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep']

# Create a dictionary to store the extracted datasets
extracted_datasets = {}

# Iterate over each column and save them separately
for column_name in columns_to_extract:
    # Extract the column values bigger than 0.1
    column_values_gt_0_1 = correlation_table.loc[correlation_table[column_name] > 0.1]

    # Reset the index
    column_values_gt_0_1.reset_index(inplace=True)

    # Extract the "DepMap_ID" column from the filtered data
    column_depmap_id = column_values_gt_0_1[['DepMap_ID']]

    # Save the "DepMap_ID" column to a CSV file
    file_name = f'/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/{column_name.replace(" ", "_")}_correlation_table_02_gt_01.csv'
    column_depmap_id.to_csv(file_name, index=False)

    # Store the extracted dataset in the dictionary
    extracted_datasets[column_name] = column_depmap_id

    # Print the saved file name
    print(f"Saved 'DepMap_ID' column for '{column_name}' to {file_name}")

# Display all the extracted datasets
for column_name, dataset in extracted_datasets.items():
    print(f"\nDataset for 'DepMap_ID' column in '{column_name}':")
    display(dataset)
```
